Comparing numpy array with python list 

```python
import numpy as np
#to bring numpy library
import time

import sys 

l =range(1000)
#list with 1000 elements
print(sys.getsizeof(5)*len(l))
#to print the size = size of one element*length of list
array=np.arange(1000)
#to create numpy array , arange =to create elements 0 to 999
print(array.size*array.itemsize)
#to print the size of array
```

Comapring numpy is more convenient than python list

```python
import numpy as np 
import sys
import time

SIZE=1000000
l1=range(SIZE)
l2=range(SIZE)

a1=np.arange(SIZE)
a2=np.arange(SIZE)

#python list
start = time.time()
result = [ (x+y) for x,y in zip(l1,l2)]
print("python list took: ", (time.time()-start)*1000)
#numpy list
start=time.time()
result = a1 + a2
print("numpy took: " , (time.time()-start)*1000)

      
```
